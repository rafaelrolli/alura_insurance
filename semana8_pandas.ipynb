```python
import pandas as pd
```

### Importando dados

```python
dados = pd.read_csv('clientes.csv', sep = ';')
dados.head(10)
```

### Exibindo a quantidade de linhas

```python
dados.shape[0]
```

### Ajustando a quantidade de sinistros

```python
dados.info()
```

```python
dados['Sinistros'].drop_duplicates()
```

```python
dados['Sinistros'].fillna(0, inplace = True)
```

```python
dados.info()
```

### Ajustando casas decimais

```python
dados[['Prêmio', 'Desconto']] = dados[['Prêmio', 'Desconto']].round(2)
dados['Sinistros'] = dados['Sinistros'].astype(int)
dados.head(10)
```

### Remoção da coluna Desconto

```python
dados.drop(columns=['Desconto'], inplace=True)
dados.head(10)
```

### Ordenando pela idade

```python
dados.sort_values(by='Idade', inplace=True, ignore_index=True)
dados
```

### Total de prêmio por estado civil

```python
total_por_estado_civil = dados.groupby('Estado Civil', as_index=False)['Prêmio'].sum().reset_index(drop=True)
total_por_estado_civil
```

### Contagem por categoria

```python
contagem_por_categoria = dados.groupby('Categoria', as_index=False)['Id'].count().rename(columns={'Id': 'Contagem'}).sort_values(by='Contagem').reset_index(drop=True)
contagem_por_categoria
```

```python
dados['Categoria'].value_counts(ascending = True)
```

# Gráficos

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (8, 4))
```

### Clientes por estado civil (gráfico de barras)

```python
valores = dados['Estado Civil'].value_counts()
rotulos = valores.index
plt.bar(rotulos, valores)
plt.title('Clientes por estado civil')
plt.xlabel('Estado civil')
plt.ylabel('Quantidade')
plt.show()
```

### Relação entre idade e prêmio (gráfico de dispersão)

```python
plt.scatter(dados['Idade'], dados['Prêmio'], s=3)
plt.title('Idade x Valor do prêmio')
plt.xlabel('Idade')
plt.ylabel('Valor do prêmio')
plt.show()
```

### Distribuição dos sinistros por sexo (gráfico de pizza)

```python
grupos = dados.groupby('Sexo')['Sinistros']
rotulos = grupos.sum().index
valores = grupos.sum().values
plt.pie(valores, labels = rotulos, autopct='%1.1f%%', startangle=45)
plt.title('Distribuição dos sinistros por sexo')
plt.show()
```

### Distribuição da quantidade de sinistros (histograma)

```python
plt.hist(dados['Sinistros'])
plt.title('Distribuição da quantidade de sinistros')
plt.xlabel('Quantidade de sinistros')
plt.ylabel('Quantidade')
plt.show()
```

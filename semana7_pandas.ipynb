### Criando a coluna "categoria"

```python
for index, row in clientes.iterrows():
  if row['idade'] >= 18 and row['idade'] <= 30:
    clientes.loc[index, 'categoria'] = 'Jovem'
  elif row['idade'] >= 31 and row['idade'] <= 45:
    clientes.loc[index, 'categoria'] = 'Adulto'
  else:
    clientes.loc[index, 'categoria'] = 'Snior'

clientes
```

```python
clientes['categoria_2'] = clientes.apply(lambda row: 'Jovem' if row['idade'] >= 18 and row['idade'] <= 30 
                     else 'Adulto' if row['idade'] >= 31 and row['idade'] <= 45
                     else 'Snior', axis = 1)
clientes
```

### Clculo do desconto

```python
def calcula_desconto(row):
  if row['sinistros'] != 0:
    return 0.0
  else:
    if row['categoria'] in ('Adulto', 'Snior'):
      return round(0.1 * row['premio'], 2)
    else:
      return round(0.05 * row['premio'], 2)

clientes['desconto'] = clientes.apply(lambda row: calcula_desconto(row), axis = 1)
clientes
```
